```python
class Empleado:
    def __init__(self, nombre, empresa, salario_base):
        self.nombre = nombre
        self.empresa = empresa
        self.salario_base = salario_base

    def calcular_salario(self):

        return self.salario_base

    def descripcion(self):
        salario_formateado = f"{self.calcular_salario():,}".replace(",", " ")
        return f"Empleado: {self.nombre} - Salario: ${salario_formateado}"


class GerenteVentas(Empleado):

    def __init__(self, nombre, empresa, salario_base, bono_ventas):
        super().__init__(nombre, empresa, salario_base)
        self.bono_ventas = bono_ventas

    def calcular_salario(self):
        return self.salario_base + self.bono_ventas

    def descripcion(self):
        salario_formateado = f"{self.calcular_salario():,}".replace(",", " ")
        return f"Gerente Ventas: {self.nombre} - Salario: ${salario_formateado}"


class ObreroPorHora(Empleado):

    def __init__(self, nombre, empresa, horas_trabajadas, valor_hora):
        super().__init__(nombre, empresa, 0)
        self.horas_trabajadas = horas_trabajadas
        self.valor_hora = valor_hora

    def calcular_salario(self):
        return self.horas_trabajadas * self.valor_hora

    def descripcion(self):
        salario_formateado = f"{self.calcular_salario():,}".replace(",", " ")
        return f"Obrero PorHora: {self.nombre} - Salario: ${salario_formateado}"



gerente = GerenteVentas("Ana Pérez", "Empresa X", 1500000, 800000)
obrero = ObreroPorHora("Pedro Soto", "Empresa Y", 160, 4000)

print(gerente.descripcion())
print(obrero.descripcion())
```

```python
from abc import ABC, abstractmethod
import math

class Figura(ABC):
    @abstractmethod
    def area(self):
        """Método abstracto para calcular el área."""
        pass

    @abstractmethod
    def perimetro(self):
        pass

    def describir(self):
        return f"Soy un(a) {self.__class__.__name__} con Área: {self.area():.2f} y Perímetro: {self.perimetro():.2f}"


class Circulo(Figura):

    def __init__(self, radio):
        self.radio = radio

    def area(self):
        return math.pi * (self.radio ** 2)

    def perimetro(self):
        return 2 * math.pi * self.radio


class Rectangulo(Figura):
    def __init__(self, base, altura):
        self.base = base
        self.altura = altura

    def area(self):
        return self.base * self.altura

    def perimetro(self):
        return 2 * (self.base + self.altura)


class Triangulo(Figura):
    def __init__(self, base, altura, hipotenusa):
        self.base = base
        self.altura = altura
        self.hipotenusa = hipotenusa

    def area(self):
        return (self.base * self.altura) / 2

    def perimetro(self):
        return self.base + self.altura + self.hipotenusa


def calcular_total(figuras: list):
    total = 0
    for figura in figuras:
        total += figura.area()
    return total


mi_circulo = Circulo(5)
mi_rectangulo = Rectangulo(4, 6)
mi_triangulo = Triangulo(3, 4, 5)

print(mi_circulo.describir())
print(mi_rectangulo.describir())
print(mi_triangulo.describir())

lista_de_figuras = [mi_circulo, mi_rectangulo, mi_triangulo]
area_total = calcular_total(lista_de_figuras)

print(f"\nSuma total de las áreas: {area_total:.2f}")
```
