```python
def filtrar_ventas(df, region, categoria, monto_minimo):
    return df[
        (df['region'] == region) &
        (df['categoria'] == categoria) &
        (df['monto_bruto'] >= monto_minimo)
    ]

filtrar_ventas(df, 'Centro', 'Laptop', 500)
```

```python
filtrar_ventas(df, 'Norte', 'Tablet', 300)
```

```python
filtrar_ventas(df, 'Sur', 'Accesorios', 100)
```

```python
resumen = df.groupby('region').agg(
    total_monto=('monto_bruto', 'sum'),
    promedio=('monto_bruto', 'mean'),
    maximo=('monto_bruto', 'max'),
    minimo=('monto_bruto', 'min'),
    conteo=('monto_bruto', 'count')
).round(2)
```

```python
resumen = resumen.reset_index().sort_values(by='total_monto')
```

```python
df_rm = df.groupby(['region', 'mes']).agg(
    total_monto=('monto_bruto', 'sum'),
    total_cantidad=('cantidad', 'sum')
).reset_index()

df_rm.head(10)
```
