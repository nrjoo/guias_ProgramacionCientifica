```python



class Producto:
    def __init__(self, nombre, precio, stock):
        self.nombre = nombre
        self.precio = float(precio)
        self.stock = int(stock)

    def __str__(self):
        return self.nombre + " - $" + str(self.precio) + " (" + str(self.stock) + " unidades)"

    def __repr__(self):
        return "Producto('" + self.nombre + "', " + str(self.precio) + ", " + str(self.stock) + ")"

    def __eq__(self, otro):
        return self.precio == otro.precio

    def __lt__(self, otro):
        return self.precio < otro.precio

    def __add__(self, otro):
        nuevo_stock = self.stock + otro.stock
        return Producto(self.nombre, self.precio, nuevo_stock)

    def __mul__(self, n):
        precio_desc = self.precio * n
        return Producto(self.nombre, precio_desc, self.stock)
```

```python
class CuentaBancaria:
    def __init__(self, titular, saldo):
        self.titular = titular
        self.saldo = saldo

    @property
    def titular(self):
        return self._titular

    @titular.setter
    def titular(self, valor):
        if len(valor) < 3:
            raise ValueError("nombre muy corto")
        self._titular = valor

    @property
    def saldo(self):
        return self._saldo

    @saldo.setter
    def saldo(self, valor):
        if valor < 0:
            raise ValueError("saldo negativo no permitido")
        self._saldo = valor

    def depositar(self, monto):
        if monto > 0:
            self.saldo = self.saldo + monto

    def retirar(self, monto):
        if monto > 0:
            self.saldo = self.saldo - monto

    def __str__(self):
        return "Cuenta de " + self.titular + " | Saldo: $" + str(self.saldo)

    def __repr__(self):
        return f"CuentaBancaria(titular='{self.titular}', saldo={self.saldo})"

    @property
    def aprobada(self):
        if self.saldo >= 10000:
            return True
        return False
```

```python
import math

class Figura:
    def __init__(self, color):
        self.color = color

    def area(self):
        raise NotImplementedError("falta hacer esto")

    def perimetro(self):
        raise NotImplementedError("falta hacer esto")

    def __str__(self):
        a = round(self.area(), 2)
        p = round(self.perimetro(), 2)
        return "Figura " + self.color + " | Área: " + str(a) + " | Perímetro: " + str(p)

    def __repr__(self):
        return self.__class__.__name__ + "('" + self.color + "')"

class Circulo(Figura):
    def __init__(self, color, radio):
        super().__init__(color)
        if radio <= 0:
            raise ValueError("radio malo")
        self.radio = radio

    def area(self):
        return math.pi * (self.radio * self.radio) # multiplicando en vez de **2

    def perimetro(self):
        return 2 * math.pi * self.radio
    
    def __repr__(self):
        return "Circulo('" + self.color + "', " + str(self.radio) + ")"

class Rectangulo(Figura):
    def __init__(self, color, ancho, alto):
        super().__init__(color)
        if ancho <= 0 or alto <= 0:
            raise ValueError("medidas malas")
        self.ancho = ancho
        self.alto = alto

    def area(self):
        return self.ancho * self.alto

    def perimetro(self):
        return 2 * (self.ancho + self.alto)

    @property
    def es_cuadrado(self):
        if self.ancho == self.alto:
            return True
        else:
            return False
    
    def __repr__(self):
        return f"Rectangulo('{self.color}', {self.ancho}, {self.alto})"
```

```python
class Inventario:
    def __init__(self):
        self._productos = []

    def agregar(self, producto):
        self._productos.append(producto)

    def __len__(self):
        return len(self._productos)

    def __getitem__(self, i):
        return self._productos[i]

    def __contains__(self, producto):
        return producto in self._productos

    def __iter__(self):
        return iter(self._productos)

    def __str__(self):
        # la forma novata de juntar strings
        texto = ""
        for p in self._productos:
            texto = texto + str(p) + "\n"
        return texto.strip() # le quito el ultimo salto de linea

    def __repr__(self):
        return "Inventario(n_productos=" + str(len(self._productos)) + ")"

    @property
    def total_valor(self):
        plata = 0
        for p in self._productos:
            plata = plata + (p.precio * p.stock)
        return plata

    @property
    def mas_caro(self):
        if len(self._productos) == 0:
            return None
        return max(self._productos)

    def buscar(self, nombre):
        lista = []
        for p in self._productos:
            if nombre.lower() in p.nombre.lower():
                lista.append(p)
        return lista
```
