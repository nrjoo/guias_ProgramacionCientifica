```python
from pathlib import Path
import pandas as pd
import numpy as np
```

```python

df = pd.read_csv('datos/ventas/ventas_enero.csv', parse_dates=['fecha'])


print('Primeras 5 filas del DataFrame:')
df.head()
```

```python
print('DIMENSIONES ')
print(f'Filas: {df.shape[0]} | Columnas: {df.shape[1]}')
print()


print(' TIPOS DE DATO')
print(df.dtypes)
print()


print('VALORES FALTANTEs')
print(df.isnull().sum())
print()

print('ESTADÍSTICAS DESCRIPTIVAS')
df.describe()
```


| Columna | Tipo de dato | Valores faltantes | Observación |

| producto | object | 0 | Categoría de texto, 8 productos distintos |
| vendedor | object | ~10% | Contiene valores nulos (vendedores sin asignar) |

```python
df_sin_nulos = df[df['vendedor'].notna()]

print(f'Total de filas originales:  {len(df)}')
print(f'Filas con vendedor válido:  {len(df_sin_nulos)}')
print(f'Filas eliminadas (nulos):   {len(df) - len(df_sin_nulos)}')
print(f'Porcentaje conservado:      {len(df_sin_nulos)/len(df)*100:.1f}%')
```

```python

carpeta = Path('datos/ventas')
archivos = sorted(carpeta.glob('*.csv'))

print(f'Archivos CSV encontrados: {len(archivos)}')
print()
for a in archivos:
    print(f'  {a.name}')
```

```python
dfs = []
filas_por_archivo = {}  

for archivo in archivos:
    df_mes = pd.read_csv(archivo, parse_dates=['fecha'])
    df_mes['mes'] = archivo.stem.replace('ventas_', '')
    filas_por_archivo[archivo.name] = len(df_mes)
    dfs.append(df_mes)

consolidado = pd.concat(dfs, ignore_index=True)

print('Consolidación completada.')
print(f'Columnas del DataFrame final: {list(consolidado.columns)}')
print(f'Forma del DataFrame consolidado: {consolidado.shape}')
consolidado.head()
```

```python
suma_filas_individuales = sum(filas_por_archivo.values())
filas_consolidado = len(consolidado)
validacion_ok = filas_consolidado == suma_filas_individuales

print(' TABLA DE VALIDACIÓN ')
print(f'Archivos CSV encontrados (glob):      {len(archivos)}')
print(f'Filas en el DataFrame consolidado:    {filas_consolidado}')
print(f'Suma de filas individuales:           {suma_filas_individuales}')
print(f'Validación exitosa:                   {"Sí " if validacion_ok else "No✗"}')
print(f'Columnas del DataFrame final:         {list(consolidado.columns)}')
print()
print('Filas por archivo:')
for nombre, filas in filas_por_archivo.items():
    print(f'  {nombre}: {filas} filas')
```

```python

orden_meses = ['enero','febrero','marzo','abril','mayo','junio',
               'julio','agosto','septiembre','octubre','noviembre','diciembre']


ventas_por_mes = consolidado.groupby('mes')['total'].sum().reindex(orden_meses)
print('Total de ventas CLP por mes:')
print(ventas_por_mes.apply(lambda x: f'${x:,.0f}').to_string())
print()

producto_mas_vendido = consolidado.groupby('producto')['cantidad'].sum().idxmax()
cantidad_total = consolidado.groupby('producto')['cantidad'].sum().max()
print(f'Producto más vendido: {producto_mas_vendido} ({cantidad_total:,} unidades)')
print()

mes_mayor_ingreso = ventas_por_mes.idxmax()
ingreso_max = ventas_por_mes.max()
print(f'Mes con mayor ingreso: {mes_mayor_ingreso} (${ingreso_max:,.0f} CLP)')
print()

promedio_transaccion = consolidado['total'].mean()
print(f'Promedio de venta por transacción: ${promedio_transaccion:,.0f} CLP')
```

```python

resumen_mensual = consolidado.groupby('mes').agg(
    total_ventas=('total', 'sum'),
    num_transacciones=('total', 'count'),
    promedio_transaccion=('total', 'mean'),
    unidades_vendidas=('cantidad', 'sum')
).reindex(orden_meses).reset_index()
resumen_mensual.columns = ['mes', 'total_ventas_clp', 'num_transacciones',
                            'promedio_transaccion', 'unidades_vendidas']

resumen_productos = consolidado.groupby('producto').agg(
    total_ventas=('total', 'sum'),
    unidades_vendidas=('cantidad', 'sum'),
    num_transacciones=('total', 'count'),
    precio_promedio=('precio_unitario', 'mean')
).reset_index().sort_values('total_ventas', ascending=False)

print('Resumen mensual:')
print(resumen_mensual.to_string(index=False))
print()
print('Resumen por producto:')
print(resumen_productos.to_string(index=False))
```

```python
Path('datos/salida').mkdir(parents=True, exist_ok=True)

ruta_csv = 'datos/salida/ventas_2024_consolidado.csv'
consolidado.to_csv(ruta_csv, index=False, encoding='utf-8')

tamano = Path(ruta_csv).stat().st_size / 1024
print(f'Archivo CSV exportado correctamente: {ruta_csv}')
print(f'Tamaño del archivo: {tamano:.1f} KB')
```

```python
# Exportar a Excel con tres hojas
ruta_excel = 'datos/salida/reporte_anual_2024.xlsx'

with pd.ExcelWriter(ruta_excel, engine='openpyxl') as writer:
    consolidado.to_excel(writer, sheet_name='Datos_Completos', index=False)
    resumen_mensual.to_excel(writer, sheet_name='Resumen_Mensual', index=False)
    resumen_productos.to_excel(writer, sheet_name='Resumen_Productos', index=False)

print(f'Archivo Excel exportado correctamente: {ruta_excel}')
with pd.ExcelFile(ruta_excel) as libro:
    print(f'Hojas en el archivo Excel: {libro.sheet_names}')
```
